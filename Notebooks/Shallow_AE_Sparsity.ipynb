```python
import numpy as np
import pandas as pd
import sys
import matplotlib.pyplot as plt
sys.path.append('/home/bastien/Documents/StageCMM/Auto-encoders/SourceCode')
import bastien_utils
from shallowAE import ShallowAE
from sparseShallowAE import SparseShallowAE_KL, SparseShallowAE_L1, SparseShallowAE_KL_sum
import datetime
```

```python
FILE_PATH = "../"
data = bastien_utils.load_data(FILE_PATH, train=True, test=True, subsetTest=True)
x_train, ytrain, x_test, y_test, x_small, y_small = data
print('x_small shape:', x_small.shape)
print('x_test shape:', x_test.shape)
```

## 1. KL_div

### 1.2 Sparsity weight = 1, sparsity paramter = 0.1

```python
AE_kl = SparseShallowAE_KL(sparsity_weight=1, sparsity_objective=0.1)
```

```python
AE_kl.train(x_train, nb_epochs=200, X_val=x_test)
```

```python
AE_kl.plot_reconstructions(x_small)
```

```python
AE_kl.plot_atoms()
```

```python
print('Training reconstruction error: ', AE_kl.reconstruction_error(x_train))
print('Test reconstruction error: ', AE_kl.reconstruction_error(x_test))
```

```python
print('Best SVM classification accuracy: ', AE_kl.best_SVM_classification_score(x_test, y_test))
```

#### 3.2 Sparsity_weights = 0.00001, Long training and early stopping

```python
from keras.callbacks import EarlyStopping
```

```python
AE_L1 = SparseShallowAE_L1(sparsity_weight=0.00001)
```

```python
AE_L1.autoencoder.fit(x_train, x_train, epochs=500, batch_size=128, callbacks=[EarlyStopping(monitor='val_loss', min_delta=0.00001, patience=10, verbose=3, mode='auto')], shuffle=True, validation_data=(x_test, x_test))
```
